```python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from torchvision import models
import numpy as np
import pandas as pd
from PIL import Image
from sklearn.model_selection import train_test_split
```

```python
# Define the UNet model
class UNet(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UNet, self).__init__()

        # Define the encoder (contracting path)
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Define the decoder (expansive path)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Define the final output layer
        self.output = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
      enc1 = self.encoder[:4](x)
      enc2 = self.encoder[4:8](enc1)
      enc3 = self.encoder[8:](enc2)

      dec1 = self.decoder[:4](enc3)
      dec2 = self.decoder[4:](dec1)
      out = self.output(dec2)
      return out
```

```python
class CustomDataset(Dataset):
    def __init__(self, meta_data_file, transform=None):
        self.meta_data = pd.read_csv(meta_data_file)
        self.transform = transform

    def __len__(self):
        return len(self.meta_data)

    def __getitem__(self, idx):
        img_name = '/kaggle/input/augmented-forest-segmentation/Forest_Segmented/Forest_Segmented/images/' + self.meta_data.iloc[idx, 0]
        mask_name = '/kaggle/input/augmented-forest-segmentation/Forest_Segmented/Forest_Segmented/masks/' + self.meta_data.iloc[idx, 1]

        image = Image.open(img_name)
        mask = Image.open(mask_name)

        mask_np = np.array(mask)
        mask = torch.tensor(mask_np, dtype=torch.long)  #        long/int

        if self.transform:
            image = self.transform(image)

        return image, mask
```

```python
#  
learning_rate = 0.001
batch_size = 16
num_epochs = 10

#  
transform = transforms.Compose([
    transforms.ToTensor()
])
```

```python
meta_data_file = '/kaggle/input/augmented-forest-segmentation/meta_data.csv'
df_meta_data = CustomDataset(meta_data_file, transform)
#       
train_data, test_data = train_test_split(df_meta_data, test_size=0.2)

train_dataset = CustomDataset(meta_data_file='/kaggle/input/augmented-forest-segmentation/meta_data.csv', transform=transform)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_dataset = CustomDataset(meta_data_file=test_data, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
#     
model = UNet(in_channels=3, out_channels=2)  # 2 :   
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
#  
for epoch in range(num_epochs):
    model.train()
    for images, masks in train_loader:
        images = images.cuda()
        masks = masks.unsqueeze(1).cuda()

        optimizer.zero_grad()
        outputs = model(images)

        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}')
```

```python
model.eval()
with torch.no_grad():
    for images, masks in test_loader:
        images = images.cuda()
        masks = masks.unsqueeze(1).cuda()

        outputs = model(images)

        #      
        loss = criterion(outputs, masks)

    print(f'Test Loss: {loss.item()}')
```
